```python
from PIL import Image

img1 = Image.open("logo1_bis.png")
img2 = Image.open("logo2_bis.png")

frames = []
steps = 1  # fade steps

for i in range(steps + 1):
    alpha = i / steps
    frame = Image.blend(img1, img2, alpha)
    frames.append(frame)

frames[0].save(
    "output_alt.gif",
    save_all=True,
    append_images=frames[1:],
    duration=1000,   # ms per frame
    loop=0
)
```
